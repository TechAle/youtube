## Indexing

```python
## Guardo tutte le colonne
studenti.columns
## Guardo la colonna nomi
studenti["Nome"]
## Un altro modo per guardare una colonna
studenti.Nome
## Guardare nome e cognomi
studenti[["Nome", "Cognome"]]
```

```python
## Prendo un dataframe usando l'index
studenti.iloc[0]
studenti.iloc[0,0]
studenti.iloc[[0,1], [0,1,2]]
```

```python
## Utilizzo senza usare l'index
studenti.loc[0, "Nome"]
studenti.loc[[0,1], ["Nome", "Cognome"]]
studenti.loc[0:2, "Nome" : "Codice"]
```
