```python
## Librerie
import pandas as pd
import numpy as np
from random import randint as rd
## Prendo il dataset
nomi = np.genfromtxt("nomi.txt", dtype="str")
## Creo un array con il numero di studenti
sezioni = np.array([11, 9, 6, 3, 2], dtype=int)
## Inizio codice utente
utente = 11981
```

```python
## Dataframe
studenti = pd.DataFrame({
                        "Nome" : [],
                        "Cognome" : [],
                        "Codice" : [],
                        "Classe" : [],
                        "Voti" : [],
                        "Promosso" : [],
                        "Comportamento" : []
                        })
## Nuovi studenti
studenti_nuovi = pd.DataFrame({
                        "Nome" : [],
                        "Cognome" : [],
                        "Codice" : [],
                        "Classe" : []
                        })
```

```python
## Itero per tutte le sezioni
for classe_num, i in enumerate(sezioni, 1):
    ## Itero per ogni classe
    for j in range(i):
        ## Genero la classe randomica
        if rd(0,1):
            tipo = "EN"
        else:
            tipo = "IN"
        ## Salvo la classe
        classe = str(classe_num) + chr(65 + j) + tipo
        ## Creo gli studenti
        for _ in range(rd(5, 10)):
            voti_l = np.random.randint(3, 11, size = (7))
            studenti = studenti.append({
                    "Nome" : np.random.choice(nomi), 
                    "Cognome" : np.random.choice(nomi),
                    "Codice" : utente,
                    "Classe" : classe,
                    "Voti" : voti_l,
                    "Promosso" : np.average(voti_l) >= 6,
                    "Comportamento" : rd(7,10)
                    }, ignore_index = True)
            utente += 1
```

```python
## Itero per ogni classe
for j in range(9):
    ## Genero la classe randomica
    if rd(0,1):
        tipo = "EN"
    else:
        tipo = "IN"
    ## Salvo la classe
    classe = '1' + chr(65 + j) + tipo
    ## Creo gli studenti
    for _ in range(rd(5, 10)):
        studenti_nuovi = studenti_nuovi.append({
                "Nome" : np.random.choice(nomi), 
                "Cognome" : np.random.choice(nomi),
                "Codice" : utente,
                "Classe" : classe
                }, ignore_index = True)
        utente += 1
```

```python
## Salvo
studenti.to_csv("studenti.csv", index = False)
studenti_nuovi.to_pickle("studenti_nuovi.pkl")
```
