## Fine anno 2/2

```python
### studenti_nuovi va bene e useremo l'output di sopra
### L'eliminazione delle quinte va bene e useremo l'output di sopra
## Modifichiamo le classi. Se l'alunno √® stato promosso
def analisi(classe, promosso):
    intero = int(classe[0])
    if promosso:
        intero += 1
    return "{}{}".format(intero, classe[1:])
copia_studenti["classe"] = copia_studenti.apply(lambda x : analisi(x["classe"], x["promosso"]), axis=1)
## Elimino tutti quelli che sono passati in quinta
filt = copia_studenti["classe"].str[0] == '6'
copia_studenti.drop(copia_studenti[filt].index, inplace = True)
## Ordiniamo per la classe
copia_studenti.sort_values(by = "classe", inplace = True)
## Elimino le colonne eccessive
copia_studenti.drop(columns = ["voti", "comportamento", "media", "promosso"], inplace = True)
copia_studenti = studenti_nuovi.append(copia_studenti)
```
