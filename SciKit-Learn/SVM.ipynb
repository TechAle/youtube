## Librerie

```python
## Grafici
import matplotlib.pyplot as plt
## Dataset disponibili (useremo digits)
from sklearn import datasets
## Algoritmo Support Vector Machine
from sklearn import svm
## Split del testo
from sklearn.model_selection import train_test_split
## Testing dei modelli
from sklearn.model_selection import GridSearchCV
## Salvataggio del file
import pickle
```

## Dati

```python
## Carichiamo il nostro dataset
dati = datasets.load_digits()
## Faccio lo split
x, x_test, target, test_target, img, img_test = train_test_split(dati["data"], dati.target, dati.images, test_size=.1)
```

## Creo il nostro algoritmo

```python
'''
    gamma = Più è alto e più curve avremo (rischio di underfitting e overfitting)
    C = Penalità di errore
'''
clf = svm.SVC(gamma = 0.1, C = 100)
## Metto i dati dentro
clf.fit(x,target)
print(clf.score(x, target))
```

## Stampe

```python
val = 0
if val <= x_test.__len__():
    print("Predizione: %s" % clf.predict(x_test[val].reshape(1, -1))[0])
    plt.imshow(img_test[val], cmap=plt.cm.gray_r)
```

## Ricerca dei parametri migliori

```python
svr = GridSearchCV(svm.SVC(), param_grid = {"C": [0.1, 1, 2],
                               "gamma": [0.1, 1, 2]})
svr.fit(x, target)
print(svr.best_params_)
```

# Quale classificatore scegliere?

* Nearest Neighbors (Supervised)
    * Molto versatile e veloce
    * Usato molto nelle comparazioni
    * Diventa estremamente lento più i dati aumentano
* Linear SVM
    * Molto versatile e veloce. Linearmente ha delle performance molto maggiori dei suoi concorrenti
    * Molto facile da applicare e trova sempre la linea migliore
    * Molto buono per le classificazioni binarie
    * Ha un O($N^3$) (consigliato avere un numero di esempi minore di $N^6$)
    * In certi ambiti (es per le immagini) non è il massimo (in questo caso è migliore il deep learning)
* RBF
    * Veloce nel training però pecca nella comparazione
* Gaussian Process
    * Molto preciso nell'output però il training è molto pesante
* Decision Tree
    * Semplice e relativamente veloce
    * Piccoli dati possono completamente stravolgere tutto quanto e si capita molto spesso nell'overfitting
* Random Forest
    * Un decision Tree più affidabile però più lento
* NeuralNet
    * Multilayer Perceptron
    * Impara in tempo reale e modelli lineari
    * Molto sensibile a una scalamento di dati
* AdaBoost
    * Aiuta molti algoritmi e ne dà una buona precisione
    * Lo squilibrio di dati può portare a un peggioramento di precisione ed è lento
* Naive Bayes
    * Lavora molto bene per la classificazione del testo ed è molto veloce
    * Presuppone che i dati siano indipendenti e che non interferiscano con altri
* QDA
    * Decisione quadratica molto veloce
    * Aumento del tempo di training a maggiori complessità
