## Librerie

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importo il dataset

```python
df = pd.read_csv( "linguaggi.csv", index_col = "Unnamed: 0")
### Prendo i dati che mi interessano
## Toglo i valori nulli
dati = df["2020"].dropna()
## ordino dal più alto al più basso
dati.sort_values(inplace = True, ascending = False)
# dati
```

## Aggiungo le percentuali

```python
## Prendo la somma di tutte le colonne
somma_totale = dati.sum()
## Creo un array vuoto che mi servirà per custodire le colonne modificate
colonne = np.empty(0, dtype = str)
## Itero
for index, value in enumerate(dati):
    ## Aggiungo mettendo in più la percentuale rispetto alla somma totale
    colonne = np.append(colonne, dati.index[index] + " {}%".format(round(value/somma_totale*100, 2)))
## Sostituisco
dati.index = colonne
# dati
```

## Creo il grafico

```python
## Utilizzo lo stile ggplot
plt.style.use("ggplot")
## Aumento le dimensioni
plt.figure(figsize=(8,10))
## Creo il grafico a barre orrizontali
plt.barh(dati.index, dati)
## Inverto l'asse delle y mettendo sopra i valori più grando
plt.gca().invert_yaxis()
## Metto il titolo
plt.title("Linguaggi 2020", fontsize = 20, color = "Black")
## Rendo il testo nell'asse delle y rosso
plt.yticks(color = "Black", fontsize = 15)
## Tolgo il testo nell'asse delle x
plt.xticks(alpha = 0)
```
