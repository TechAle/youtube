## Librerie

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Creiamo dataset

```python
## Creo il nostro range di dati
x = np.array([x/100 for x in range(700)], dtype = float) 
## Prendo il seno
y_sin = np.sin(x)
## Prendo il coseno
y_cos = np.cos(x)
```

## Grafici

```python
## Creo il subplot
fig, (x1, x2) = plt.subplots(nrows = 2, ncols = 1)
# fig1, x1 = plt.subplots()
# fig2, x2 = plt.subplots()
'''
    Configurazione:
    prima riga -> coseno
    seconda riga -> seno
'''
## Metto i vari dati nei grafici
x1.plot(x, y_cos)
x2.plot(x, y_sin)
## Setto il label per l'asse delle y
x1.set_ylabel("Coseno")
x2.set_ylabel("Seno")
## Setto l'asse delle X solo per il seno
x2.set_xlabel("Tempo")
## Dividiamo per bene i vari subplot
plt.tight_layout()
## Mostro
plt.show()
```
