## Librerie

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Importo dataset

```python
df = pd.read_csv( "linguaggi.csv", index_col = "Unnamed: 0")
df.fillna(0, inplace = True)
```

## Preparo i dati

```python
## Asse x
ascissa = df.columns
## Linguaggi che vogliamo vedere
linguaggi_voluti = ["Python", "Java"]
```

```python
## Setto uno stile
# print(plt.style.available)
plt.style.use("seaborn-whitegrid")
## Inserisco i dati
for i in linguaggi_voluti:
    plt.plot(ascissa, df.loc[i], label = i, marker = ".")
## Ci aggiungo c
plt.plot(ascissa, df.loc["C"], label = "C", marker = ".", linestyle = "--", color = 'r' )
## Attivo i label
plt.legend()
## Aggiungo il titolo
plt.title('Confronto Python JavaScript')
## Metto un nome agli assi
plt.xlabel('Anni')
plt.ylabel('Linguaggi')
## Mostro il grafico
plt.show()
```
